## Starting operations

Convert some variables: Travel time and inicial hour

```python
df = pd.read_csv("data_proyect.csv")
df["hora_f"] = pd.to_datetime(df["hora_f"])
df["hora_i"] = pd.to_datetime(df["hora_i"])
df["travel_time"] = df["hora_f"] - df["hora_i"]
df["travel_time"] = [i.seconds//3600+1 for i in df["travel_time"]]
df["inicial_hour"] = [i.hour for i in df["hora_i"]]
df["final_hour"] = df["inicial_hour"] + df["travel_time"]
df
```

Let's extract the geolocation of each airport and add it to the dataframe

```python
# Let's open the file
codes = pd.read_csv("iata-icao.csv", sep=";")

# Make de diccionary
dic_codes = {row["iata"]:[row["latitude"],row["longitude"]] for i,row in codes.iterrows()}

# Add the geolocations to the df
l = []
for i in df["code_i"]:
    if i == "GRU":
        l.append([-23.435555, -46.473055])
    elif i == "YEG":
        l.append([53.30871, -113.58068])
    elif i == "GMP":
        l.append([37.55749, 126.800496])
    elif i == "GIG":
        l.append([-22.805330, -43.239499])
    else:
        l.append(dic_codes[i])
df["coords_i"] = l

l2 = []
for i in df["code_f"]:
    if i == "GRU":
        l2.append([-23.435555, -46.473055])
    elif i == "YEG":
        l2.append([53.30871, -113.58068])
    elif i == "GMP":
        l2.append([37.55749, 126.800496])
    elif i == "GIG":
        l2.append([-22.805330, -43.239499])
    elif i == "CTS":
        l2.append([42.771830, 141.688830])
    else:
        l2.append(dic_codes[i])
df["coords_f"] = l2
df
```

We extract the minimum and the maximum hour to add new_variables

```python
MIN = np.min(df["inicial_hour"])
MAX = np.max(df["final_hour"])
for i in range(MIN, MAX+1):
    st = f"hour{i}"
    df[st] = " "
df
```

We change format of coordinates

```python
for i,row in df.iterrows():
    try:
        ini = [row["coords_i"][0].replace(",","."), row["coords_i"][1].replace(",",".")]
    except:
        pass
    try:
        fin = [row["coords_f"][0].replace(",","."), row["coords_f"][1].replace(",",".")]
    except:
        pass
    ini[0], ini[1] = float(ini[0]), float(ini[1])   
    fin[0], fin[1] = float(fin[0]), float(fin[1])   
    df["coords_i"][i] = ini
    df["coords_f"][i] = fin
```
